```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import chromedriver_binary
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
browser = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
browser.get('https://www.youtube.com/c/GeeksforGeeksVideos/videos')
```

```python
 

soup = BeautifulSoup(browser.page_source, 'html.parser')

complete_video_details = []
 
for video in soup.find_all(class_='ytd-rich-item-renderer'):
    try:
         
        video_description = video.find('a', class_='yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media')
        video_description = video_description.text.strip() if video_description else None
    except Exception as e:
        video_description = None 

    try:
       
        video_meta = video.find_all('span', class_='inline-metadata-item style-scope ytd-video-meta-block')
        video_views = video_meta[0].text.strip() if len(video_meta) > 0 else None
    except Exception as e:
        video_views = None 
    
    try:
        
        video_date = video_meta[1].text.strip() if len(video_meta) > 1 else None
    except Exception as e:
        video_date = None 

    try:
       
        video_url = 'https://www.youtube.com' + video.find_all('a')[0].get('href')
    except Exception as e:
        video_url = None 
    
    
    complete_video_details.append({
        'description': video_description,
        'views': video_views,
        'date': video_date,
        'url': video_url
    })

 
# for detail in complete_video_details:
#     print(detail)


df = pd.DataFrame(complete_video_details)
# Print the DataFrame
df = df.dropna()
df = df.drop_duplicates()

print(df)
df.to_csv("index1.csv", index=False)
```

# Automating 


```python
import time
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# Initialize the browser
browser = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Initialize a list to store all video details
complete_video_details = []

for video_url in df['url']:  # Iterate over the URLs in your DataFrame
    # Convert video_url to string in case it's not
    video_url = str(video_url)

    # Open the video URL in the browser
    browser.get(video_url)
    
    # Wait for the page to load
    time.sleep(2)  # Sleep for 2 seconds between requests

    # Get the page source after loading
    soup = BeautifulSoup(browser.page_source, 'html.parser')

    # Extract video title
    oneVideo = soup.find_all('yt-formatted-string', class_='style-scope ytd-watch-metadata')
    video_title = oneVideo[0].text if oneVideo else None

    # Extract video views
    try:
        videoViews = soup.find('span', class_='bold style-scope yt-formatted-string')
        video_views = videoViews.text if videoViews else None
    except Exception as e:
        video_views = None
        print(f"Error extracting views: {e}")

    # Extract video descriptions
    description_list = []
    span = soup.find_all('span', class_='yt-core-attributed-string--link-inherit-color')
    for desc in span:
        if desc:
            description_list.append(desc.text.strip())

    # Concatenate all descriptions into one string
    full_description = " ".join(description_list)

    # Extract video IDs
    video_id_list = []
    try:
        video = soup.find_all('a', class_='yt-simple-endpoint style-scope yt-formatted-string bold')
        for vid in video:
            video_id_list.append(vid.get('href') if vid.get('href') else None)
    except Exception as e:
        print(f"Error extracting video IDs: {e}")
    concatenated_video_ids = "".join(video_id_list)

    # Append the extracted information to the complete_video_details list
    complete_video_details.append({
        'title': video_title,
        'views': video_views,
        'description': full_description,  # Concatenated description
        'video_id': concatenated_video_ids,  # Use concatenated_video_ids here
        'url': video_url  # Add the video URL for reference
    })
    print(complete_video_details)
    print("*" * 40)

# Print the complete video details

```

```python
video_details_df = pd.DataFrame(complete_video_details)
 
video_details_df.to_csv("analysis.csv", index=False)
```
