```python
df["Height"].describe()
```

```python
max_limit = df["Height"].quantile(0.99)
min_limit = df["Height"].quantile(0.01)

# max_limit = df["Height"].quantile(0.99):

# This calculates the 99th percentile of the "Height" column. The 99th percentile is the value below which 99% of the data falls. This represents the upper threshold for your data, meaning values above this will be considered outliers.
# min_limit = df["Height"].quantile(0.01):

# This calculates the 1st percentile of the "Height" column. The 1st percentile is the value below which 1% of the data falls. This represents the lower threshold for your data, meaning values below this will also be considered outliers.
# new_data = df[(df["Height"] > min_limit) & (df["Height"] < max_limit)]:

# This step filters the original dataframe df to create a new dataframe new_data by removing the outliers.
# It keeps only the rows where the height is between the 1st and 99th percentiles.
# Heights greater than the 99th percentile (above max_limit) are removed.
# Heights less than the 1st percentile (below min_limit) are also removed
```

```python
max_limit
```

```python
# Trimming
new_data = df[(df["Height"]>min_limit) & (df["Height"]<max_limit)]
```

```python
sns.distplot(new_data["Height"])
```

```python
sns.boxplot(new_data["Height"])
```

```python
#capping
new_data2 = df.copy()
new_data2["Height"] = np.where(df["Height"]>max_limit,max_limit, np.where(df["Height"]<min_limit,min_limit, df["Height"]))
```

```python
sns.distplot(new_data2["Height"])
```
