# Generating Sin and Cos Graph

```python
x = np.linspace(0,np.pi, 30)
xz= np.linspace(0,4.5*np.pi, 30)
sinzx = np.sin(xz)
print(x )
print("*"*40)
sinx = np.sin(x)
print([i for i in sinx])

plt.title("Sine wave")
plt.plot(xz,sinzx )
plt.show()
plt.title("Sine wave")
plt.plot(xz,sinx )
```

```python
cosx = np.cos(x)
coxz =np.cos(xz)
plt.title("cos wave")
plt.plot(xz,sinzx,label='cos 4.5pi' )

plt.title("cos wave")
plt.plot(xz,cosx, label ='Cos pi' )

plt.legend(loc= 'upper right')
plt.title('Cos wave')
plt.show()
```

# Patterns 


```python
cosx = np.cos(x)
coxz =np.cos(xz)
plt.title("cos wave")
plt.plot(xz,sinzx,"*")

plt.title("cos wave")
plt.plot(xz,cosx,"-",linewidth = '8' )
 
plt.show()
```

# Pie Charts

# Basic operations in pie chart 

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   10  ,  200   , 500      ]
explode = [0, 0, 0 , 0, .1, 0] 
fig = plt.figure(figsize=(8,3))
fig = plt.pie(data,labels=cars, explode=explode)
plt.show()
```

```python
# Adding shadows 
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   10  ,  200   , 500      ]
explode = [0, 0, 0 , 0, .1, 0] 
fig = plt.figure(figsize = (7,7))
plt.pie(data, labels = cars, explode = explode, shadow = True)

plt.show()
```

```python
## Adding percentages 
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   50  ,  200   , 500      ]
explode = [0, 0, 0 , 0, .1, 0] 

fig = plt.figure(figsize = (7,7))

plt.pie(data, labels = cars, explode = explode, shadow = True, autopct = '%2.2f%%')
plt.show()
```

# Additiaonl styling 

```python
fig = plt.figure(figsize = (7,7))

plt.pie(data, labels = cars, explode = explode, shadow = True, autopct = '%2.2f%%',
        wedgeprops = {'edgecolor' : 'black', 'linewidth' : 2, 'antialiased' : True})

plt.show()
```

# Multiple subplots 

```python
x = np.linspace(0,10,50)

fig , ax = plt.subplots(2,2)
#fig is the figure object that contains all the plots.
# ax is a 2D array (or matrix) of axes objects that represent each individual subplot.
y = np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('Sin Wave')


y = np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title('Cos Wave')

y = np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title('Tan Wave')

y = np.random.randn(1000)
ax[1,1].hist(y,bins = 40)
ax[1,1].set_title('Random Distribution')


plt.tight_layout()
plt.show()
```
