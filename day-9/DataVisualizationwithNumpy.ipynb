```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4 * np.pi, 100)
print(x)

y = np.sin(x)
cosy = np.cos(x)
plt.plot(x,y,"o",linestyle="--", color = 'red', markersize = 5,alpha = .45)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title('Basic Line Plot ')
plt.plot(x, y, "o", linestyle="--", color='red', markersize=5, alpha=0.9, label="sin(x)")
plt.plot(x, cosy, "o", linestyle="--", color='blue', markersize=5, alpha=0.9, label="cos(x)")
plt.legend() 
plt.plot()

plt.plot(x,cosy,"o",linestyle="--", color = 'blue', markersize = 5,alpha = .33)
plt.plot()
```

# Scatter Plot

```python
x = np.random.rand(500)
y = np.random.rand(500)  
z=np.random.rand(500) # used for colors 

x1 = np.random.rand(200) + .5
y1 = np.random.rand(200) + .5  # showing this scatter thing as +2 in x axis to differenntiate



plt.scatter(x1,y1, marker = 'o', alpha = .7)
plt.scatter(x,y,marker='o' , c=z , cmap='viridis')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Basic Scatte Plot ')
plt.legend()
plt.colorbar()
plt.show()
```

#  Hexbinning Scatter Plot


```python
x1 = np.random.rand(10000)
y1 = np.random.rand(10000)
plt.hexbin(x1,y1,gridsize= 20) 

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Hexbining Plot')
plt.colorbar()
plt.show()
```

# Grouped Bar Graph 

```python
data = np.array([
    [10,20],
    [40,50]
])

width = 0.2
x= np.arange(2)

plt.bar(x-width , data[:,0] , width, label  = 'D-1')
plt.bar(x , data[:,1] , width, label  = 'D-2')
```

#  Histogram

```python
data1 = np.random.randn(100000)
data2 = np.random.randn(100000) + 2
 # bins is the no of intervals 
plt.hist(data1, bins = 100, alpha = .7, label = 'D1', color='red')
plt.hist(data2, bins = 100, alpha = .7, label = 'D2')
plt.legend()
plt.show()
```

# Cumulative Histogram

```python
data = np.random.randn(1000000)

plt.hist(data, bins = 1000, cumulative = True)
plt.show()
```

# Pie chart 

```python
labels = ['Python', 'C++', 'Java', 'JavaScript', 'PHP']
sizes =  [21.3,        14.6 , 9.2,     8.5,       7.7]
explode = (0.13,        .1,    0,       0,           0)
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightpink']


plt.title("Programming Language Popularity")
# decimal palces is autopct 
plt.pie(sizes, colors=colors, labels=labels, autopct='%1.1f%%', shadow=True, explode=explode ,startangle=140)
plt.show()
```
